# Building a dashboard with Plotly and Streamlit

## Dependencies

```python
import time
import numpy as np
import pandas as pd
import plotly.express as px
import streamlit as st

#@st.experimental_memo
def read_dataframe(input_data: str) -> pd.DataFrame:
    return pd.read_csv(input_data)

def main():
    # input
    input_data = "https://raw.githubusercontent.com/Lexie88rus/bank-marketing-analysis/master/bank.csv"
    
    # 1. read the data
    df = read_dataframe(input_data)
    
    # 2. basic dashboard setup
    st.set_page_config(
        page_title="Real-Time Data Science Dashboard",
        page_icon="âœ…",
        layout="wide",
    )
    
    return df

df = main()
df.sample(5)
```
